```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load data
df = pd.read_csv("students_cleaned.csv")

# Data overview
print(df.head())
print("Total students:", len(df))
print(df.info())

# 2. Basic stats
print("Highest Marks:", df['Marks'].max())
print("Lowest Marks:", df['Marks'].min())
print("Average Age:", df['Age'].mean())
print("Grade counts:\n", df['Grade'].value_counts())

# 3. Filtering
grade_A = df[df['Grade'] == 'A']
delhi_students = df[df['City'] == 'Delhi']
marks_above_80 = df[df['Marks'] > 80]

# 4. Sorting
print(df.sort_values(by='Age', ascending=False))
print(df.sort_values(by='Marks', ascending=True))

# 5. Visualization
df['Grade'].value_counts().plot(kind='bar', title='Students per Grade')
plt.show()

df['Marks'].plot(kind='hist', bins=5, title='Marks Distribution')
plt.show()

df['City'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.show()

# 6. Save filtered data
grade_A.to_csv("grade_A.csv", index=False)
df['City'].value_counts().to_csv("city_counts.csv")
```
