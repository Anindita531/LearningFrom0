```python
import pandas as pd

# Load data with proper delimiter and quoting
data = pd.read_csv("hw_200.csv", delimiter=",", quotechar='"', skipinitialspace=True)

# Clean column names
data.columns = data.columns.str.strip().str.replace('"', '')

# Show actual column names
print("Columns:", data.columns.tolist())

# Show first 5 rows
print(data.head())

# Summary statistics
print(data.describe())

# Average height
print("Average Height:", data['Height(Inches)'].mean())

# Tallest person
print("Max Height:", data['Height(Inches)'].max())
```

```python
import pandas as pd

df = pd.read_csv("students.csv")
print(df.head())

# Fill missing marks with average
df['Marks'].fillna(df['Marks'].mean(), inplace=True)

# Drop rows where 'City' is missing
df.dropna(subset=['City'], inplace=True)

# Fill missing Age with median
df['Age'].fillna(df['Age'].median(), inplace=True)
```
