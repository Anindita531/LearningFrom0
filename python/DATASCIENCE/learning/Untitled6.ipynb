```python
import pandas as pd

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
df = pd.read_csv(url, sep=';')  # Note: this file uses semicolon as delimiter
print(df.head())
```

```python
#Find  with alcohol > 12% and quality >= 7.
fillteree1=df[(df['alcohol']>1.2)&(df['quality']>=7)]
print(fillteree1)
```

```python
#Group by quality and find average pH.
print(df.groupby('quality')['pH'].mean())
```
