```python
import pandas as pd

# Create sample data
data = {
    'Name': ['Aditi', 'Rahul', 'Sneha', 'Arjun', 'Priya'],
    'Age': [20, 21, 19, 22, None],
    'Marks': [85, None, 90, 70, 95],
    'City': ['Kolkata', 'Mumbai', 'Delhi', None, 'Chennai']
}

df = pd.DataFrame(data)

# Show first few rows
print(df)
```

```python

#Step 4 — Students with marks > 80
# Show first few rows
print(df)
print("Students with marks > 80:")
print(df[df['Marks'] > 80])
```

```python

#Step 5 — Handle missing data (fill with mean)
#Instead of your old inplace=True method, we’ll do:
df['Marks'] = df['Marks'].fillna(df['Marks'].mean())
df['Age'] = df['Age'].fillna(df['Age'].median())
df['City'] = df['City'].fillna("Unknown")

print("After filling missing values:")
print(df)
```

```python
import matplotlib.pyplot as plt

plt.hist(df['Marks'], bins=5, edgecolor='black')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.title('Distribution of Marks')
plt.show()
```

```python

df = pd.DataFrame(data)
#Task 2 — Basic Analysis
#Find the average age of students.
print("Average age of students:",df['Age'].mean())

#Find the maximum marks scored.
print("Maximum marks scored :",df['Marks'].max())

#List all students from Delhi.
print(df[df['City']=='Delhi'])
```

```python

#Task 3 — Filtering
#Show all students with marks > 80.
print("All students with marks > 80 :\n",df[df['Marks']>80])

#Show students who are older than 20 and have marks less than 90.

print("All students who are older than 20 and have marks less than 90. :\n",df[(df['Marks']<90 )&(df['Age']>20)])
```

```python

#Task 4 — Missing Data Handling
#Count how many missing values are there in each column.
print(df.isnull().sum())
#Fill missing Marks with the mean marks.
df['Marks'].fillna(df['Marks'].mean(), inplace=True)
print(df)
#Fill missing Age with the median age.
df['Age'].fillna(df['Age'].median(), inplace=True)
print(df)

#Fill missing City with "Unknown"
df['City'].fillna("Unknown", inplace=True)
print(df)
```

```python

 
#Task 5 — Visualization
#Draw a histogram of marks.
import matplotlib.pyplot as plt
plt.hist(df['Marks'],bins=5,edgecolor='black')

plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.title('Distribution of Marks')
plt.show()
#Draw a bar chart showing average marks for each city.

# Group by city and calculate average marks
avg_marks_per_city = df.groupby('City')['Marks'].mean()

# Plot bar chart
plt.figure(figsize=(8, 5))
plt.bar(avg_marks_per_city.index, avg_marks_per_city.values )
plt.xlabel('City')
plt.ylabel('Average Marks')
plt.title('Average Marks per City')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

def assign_grade(marks):
    if marks >= 85:
        return "A"
    elif marks >= 70:
        return "B"
    else:
        return "C"

df['Grade'] = df['Marks'].apply(assign_grade)
print(df)
```

```python
df = df.sort_values(by='Marks', ascending=False)
print(df)
```

```python
df.to_csv("students_cleaned.csv", index=False)
print("Cleaned data saved to students_cleaned.csv")
```
