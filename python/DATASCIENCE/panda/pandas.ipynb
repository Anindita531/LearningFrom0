```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['Delhi', 'Mumbai', 'Kolkata']
}
#2. DataFrame - ২D টেবুলার ডেটা স্ট্রাকচার
df = pd.DataFrame(data)
print(df)
s = pd.Series(data)
print(s)
```

```python
# ডেটা ফিল্টার করা
print("------------Age greater than 28-------------")
print(df[df['Age'] > 28])
print("-------------------------")
# নতুন কলাম যোগ করা
print("------------Age in 5 years -------------")
df['Age_in_5_years'] = df['Age'] + 5
print(df)
print("-------------------------")
# কলাম ডিলিট করা
print("------------delete-------------")
df = df.drop('Age_in_5_years', axis=1)
print("-------------------------")
# গ্রুপ বাই করা
print("-----------column wise sort to calculate mean -------------")
print(df.groupby('City')['Age'].mean())
print("-------------------------")
# ডেটা ফিল্টার করা
print(df[df['Age'] > 28])
print("-------------------------")
# নতুন কলাম যোগ করা
df['Age_in_5_years'] = df['Age'] + 5
print(df)
print("-------------------------")
# কলাম ডিলিট করা
df = df.drop('Age_in_5_years', axis=1)
print("-------------------------")
# গ্রুপ বাই করা
print(df.groupby('City')['Age'].mean())
```

```python
df2 = pd.DataFrame({
    'Name': ['Raju', 'Mina', 'Raju'],
    'Age': [23, 27, 23]
})

print("Before dropping duplicates:")
print(df2)

df2 = df2.drop_duplicates()
print("\nAfter dropping duplicates:")
print(df2)
```

```python
 
# First DataFrame
df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie']
})

# Second DataFrame
df2 = pd.DataFrame({
    'ID': [1, 2, 4],
    'Score': [85, 90, 75]
})

# Merge on 'ID'
merged_df = pd.merge(df1, df2, on='ID', how='inner')
print(merged_df)
```

```python
# First DataFrame
df_a = pd.DataFrame({
    'Name': ['Alice', 'Bob'],
    'Score': [85, 90]
})

# Second DataFrame
df_b = pd.DataFrame({
    'Name': ['Charlie', 'David'],
    'Score': [75, 88]
})

# Concatenate vertically
concat_df = pd.concat([df_a, df_b])
print(concat_df)
```
