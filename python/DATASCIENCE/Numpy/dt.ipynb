```python
from sklearn.tree import DecisionTreeClassifier

# Features: Complexity (0=Low,1=Medium,2=High), Time
X = [[0,1],[1,2],[2,1],[2,3],[1,4]]
y = [0,0,1,1,0]  # 0=No, 1=Yes

clf = DecisionTreeClassifier(criterion='entropy')
clf.fit(X, y)

# Predict new task
print(clf.predict([[2,2]]))  # Output: [1] => High Priority
```
