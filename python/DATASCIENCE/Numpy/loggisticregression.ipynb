```python
from sklearn.linear_model import LogisticRegression
import numpy as np

# Features: [Hours Studied, Hours Slept]
X = np.array([
    [1, 5],
    [2, 6],
    [3, 5],
    [4, 6],
    [5, 7]
])
y = np.array([0, 0, 1, 1, 1])

model = LogisticRegression()
model.fit(X, y)

pred = model.predict([[3, 6]])
print("Prediction:", "Pass" if pred[0] == 1 else "Fail")
print("Probability:", model.predict_proba([[3, 6]])[0])
```
