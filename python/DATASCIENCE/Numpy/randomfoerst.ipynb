```python
from sklearn.ensemble import RandomForestRegressor
import numpy as np

# Features: [Complexity, Team Size]
X = np.array([
    [1, 2],
    [2, 2],
    [3, 4],
    [4, 5],
    [5, 7]
])
y = np.array([3, 4, 6, 8, 12])

model = RandomForestRegressor(n_estimators=3, random_state=42)
model.fit(X, y)

prediction = model.predict([[3, 3]])  # new task
print("Predicted Time (hrs):", prediction[0])
```
