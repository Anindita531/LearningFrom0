```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#৫. ডেটাসেট লোড করা (Seaborn এর বিল্ট-ইন ডেটাসেট)
df = sns.load_dataset('tips')  # tips dataset লোড করা (popular dataset)
print(df.head())
```

```python
#\1 ৬.১ scatterplot (বিন্দু ছড়িয়ে দেওয়া)
sns.scatterplot(data=df, x='total_bill', y='tip', hue='day')
plt.show()
```

```python
#৬.২ lineplot (রেখাচিত্র)
sns.lineplot(data=df, x='size', y='total_bill')
plt.show()
```

```python
#৬.৩ barplot (বার চার্ট)
sns.barplot(data=df, x='day', y='total_bill')
plt.show()
```

```python
#৬.৪ countplot (categorical count)
sns.countplot(data=df, x='day')
plt.show()
```

```python
#৬.৫ histplot (Histogram)
sns.histplot(data=df, x='total_bill', bins=10)
plt.show()
```

```python
#৬.৬ boxplot (বক্স প্লট)

sns.boxplot(data=df, x='day', y='total_bill')
plt.show()
```

```python
#৬.৭ violinplot (ভায়োলিন প্লট)

sns.violinplot(data=df, x='day', y='total_bill')
plt.show()
```

```python
#৬.৮ heatmap (হিটম্যাপ)

import numpy as np

numeric_df = df.select_dtypes(include=[np.number])
size_corr = numeric_df.corr()['size'].sort_values(ascending=False).to_frame()

sns.heatmap(size_corr, annot=True, cmap='viridis', cbar=False)
plt.ylabel('Feature')
plt.title('Correlation with size')
plt.show()
```

```python
#৬.৯ pairplot (পিয়ারপ্লট)

sns.pairplot(df)
plt.show()
```

```python
sns.jointplot(data=df, x='total_bill', y='tip', kind='scatter')
plt.show()

```

```python
sns.catplot(data=df, x='day', y='total_bill', kind='box')
plt.show()
```

```python
sns.set_style('darkgrid')  # options: darkgrid, whitegrid, dark, white, ticks
sns.set_context('notebook')  # options: paper, notebook, talk, poster
```

```python
#full
import seaborn as sns
import matplotlib.pyplot as plt

# ডেটাসেট লোড
df = sns.load_dataset('tips')

# স্টাইল সেট
sns.set_style('whitegrid')

# বার প্লট
plt.figure(figsize=(8,5))
sns.barplot(data=df, x='day', y='total_bill', palette='pastel')
plt.title('Total Bill per Day')
plt.show()

# বক্স প্লট
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='day', y='tip', hue='sex')
plt.title('Tip by Day and Gender')
plt.show()

# হিটম্যাপ
plt.figure(figsize=(7,5))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
